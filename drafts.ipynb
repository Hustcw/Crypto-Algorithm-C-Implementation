```python
def circular_left(b, r):
    return (((b << r) & 0xff) | (b >> (8-r)))&0xff

def affine(b):
    return b ^ circular_left(b,1) ^ circular_left(b,2) ^ circular_left(b,3) ^ circular_left(b, 4) ^ 0x63
```

```python
s_box = [0] * 256    
p = q = 1
while 1:
    p = (p ^ (p << 1) ^ (0x11B if(p&0x80) else 0))&0xff
    q ^= q << 1;
    q ^= q << 2;
    q ^= q << 4;
    q ^= (0x09 if (q & 0x80) else 0)
    q &= 0xff
    s_box[p] = hex(affine(q))
    if(p==1): break
s_box[0] = hex(0x63)
for i in range(16):
    print(s_box[i*16:i*16+16])
```
